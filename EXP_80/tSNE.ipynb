```python
import numpy as np
import pandas as pd
from openTSNE import TSNE
import matplotlib.pyplot as plt

%matplotlib inline
```

```python
!ls
```

```python
data = np.load('adap_layer_pred.npy')
df = pd.read_csv('val_resuls.csv')
```

```python
embedding = TSNE(n_jobs=-1, perplexity=100, learning_rate=100).fit(data)
```

```python
fig, axes = plt.subplots(figsize = (10, 10))
axes.scatter(embedding[:, 0], embedding[:, 1], alpha =0.01, c=df['vowel_diacritic'])
```
